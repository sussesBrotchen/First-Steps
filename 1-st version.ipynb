### Задание №1
* Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

```python
phrase_1 = 'Хочу съесть целый торт, но уже час ночи.'
phrase_2 = 'Я питаюсь только энергией космоса.'

a = int(len(phrase_1))
b = int(len(phrase_2))

if a > b:
    print('Результат - Фраза 1 длиннее фразы 2')
elif a < b:
    print('Результат - Фраза 2 длиннее фразы 1')
else:
    print('Результат - Фбе фразы равны')
```

### Задание №2
* Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

```python
user_year = int(input('Введите год - '))

if user_year % 4 == 0:
    print('Результат:')
    print('Это високосный год.')
else:
    print('Результат:')
    print('Этот год не является високосным.')
```
