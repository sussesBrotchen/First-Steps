### Задание №1
* Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

```python
phrase_1 = 'Хочу съесть целый торт, но уже час ночи.'
phrase_2 = 'Я питаюсь только энергией космоса.'

a = int(len(phrase_1))
b = int(len(phrase_2))

if a > b:
    print('Результат - Фраза 1 длиннее фразы 2')
elif a < b:
    print('Результат - Фраза 2 длиннее фразы 1')
else:
    print('Результат - Фбе фразы равны')
```

### Задание №6
* Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

```python
triangle = ['Треугольник', 'треугольник']
square = ['Прямоугольник', 'прямоугольник']
circle = ['Круг', 'круг']

user_input_shape = input('Укажите фигуру (круг, прямоугольник или треугольник): ')

if user_input_shape in square:
    square_width = input('Укажите ширину: ')
    if square_width.isdigit():
        s1 = int(square_width)
        square_length = input('Укажите длину: ')
        if square_length.isdigit():
            s2 = int(square_length)
            print(f'Площадь прямоугольника: {s1 * s2}')
        else:
            print('Ошибка!')
    else:
        print('Ошибка!')
elif user_input_shape in circle:
    circle_r = input('Укажите радиус: ')
    if circle_r.isdigit():
        r = int(circle_r)
        print(f'Площадь круга: {3.14 * r ** 2}')
    else:
        print('Ошибка!')
elif user_input_shape in triangle:
    tri_a = input('Введите сторону А: ')
    tri_b = input('Введите сторону B: ')
    tri_c = input('Введите сторону C: ')
    if tri_a.isdigit() and tri_b.isdigit() and tri_c.isdigit():
        a = int(tri_a)
        b = int(tri_b)
        c = int(tri_c)
        p = (a + b + c) / 2
        print(f'Площадь треугольника: {(p * ((p - a) * (p - b) * (p - c))) ** 0.5}')
    else:
        print('Ошибка!')
else:
    print('Кажется, где-то спряталась ошибка.')
```

### Задание №7
* Составьте калькулятор для вычисления возраста собак и кошек с человеческих лет.

```python
# Calculator for determine your pet age


animal_type = [
                'Собака',
                'собака',
                'Собаки',
                'собаки',
                'Dog',
                'dog',
                'Кошка',
                'кошка',
                'Кошки',
                'кошки',
                'Cat',
                'cat'
               ]


print('Привет! Давайте определим возвраст вашего питомца!')

print('Мы можем посчитать возраст собаки или кошки.')


user_input_animal_type = input('Чей возраст вы хотите узнать? ')


if user_input_animal_type in animal_type:
    pet_age = input('Укажите возраст питомца в человеческих годах: ')


    if pet_age.isdigit():
        pet_age = int(pet_age)

        if pet_age == 1:
            print('Вашему питомцу 15 лет')

        elif pet_age == 2:
            print('Вашему питомцу 24 года')
        
        elif pet_age > 2:

            # Формула: 15 (1-й год) + 9 (2-й год) + (возраст-2)*4 (последующие годы)

            pet_years = 15 + 9 + (pet_age - 2) * 4
            
            print(f'Вашему питомцу {pet_years} лет')

        else:
            print('Кажется, что-то пошло не так.')


    else:
        print('Возраст должен быть указан цифрами и быть больше нуля.')


else:
    print('Допущена ошибка при вводе типа питомца.')
```
